```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/HP/OneDrive/Desktop/Data Analysis/company sales dataset/company_sales_data.csv")
df.head()
```

```python
# Exercise 9: Read Bathing soap, facewash of all months and display it using the Subplot.
fig,(ax1, ax2) = plt.subplots(2 , 1 , sharex=True)
ax1.plot(df["month_number"] , df["bathingsoap"] , '-g' , label='Bathing Soap')
ax1.set_ylabel('Quantity')
ax1.set_title('Monthly Consumption')
ax1.legend()
ax2.plot(df["month_number"] , df["facewash"] , '-b' , label='Face Wash')
ax2.set_xlabel('Month')
ax2.set_ylabel('Quantity')
ax2.legend()
```

```python
# Exercise 10: Read all product sales data and show it using the stack plot.
plt.stackplot(df["month_number"] , df["facecream"] , df["facewash"] , df["toothpaste"] , df["bathingsoap"] , df["shampoo"] , df["moisturizer"] , labels = ("facecream" , "facewash" , "toothpaste" , "bathingsoap" , "shampoo" , "moisturizer") , colors=['m','c','r','k','g','y'])
plt.legend(loc = "upper left")
plt.xlabel("month number")
plt.ylabel("product sales data")
```
